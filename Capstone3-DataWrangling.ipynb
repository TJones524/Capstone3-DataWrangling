```python
loans['Status_Checking_Acc'] = loans['Status_Checking_Acc'].replace(['A11', 'A12', 'A13', 'A14' ], ['Neg', 'Low', 'High', 'NoAcc'])
```

```python
loans['Credit_History'] = loans['Credit_History'].replace(['A30', 'A31', 'A32', 'A33', 'A34' ], ['NoCredit', 'Paid', 'Current', 'Delay', 'Critical'])
```

```python
loans['Purpose_Credit_Taken'] = loans['Purpose_Credit_Taken'].replace(['A40', 'A41', 'A42', 'A43', 'A44', 'A45', 'A46', 'A47', 'A48', 'A49', 'A410' ], ['NewCar', 'UsedCar', 'Furniture', 'Radio_TV', 'Appliances', 'Repairs', 'Education', 'Vacation', 'Retrain', 'Business', 'Other'])
```

```python
loans['Savings_Acc'] = loans['Savings_Acc'].replace(['A61', 'A62', 'A63', 'A64', 'A65'], ['Low', 'LowAvg', 'Avg', 'High', 'NoAcc'])
```

```python
loans['Years_At_Present_Employment'] = loans['Years_At_Present_Employment'].replace(['A71', 'A72', 'A73', 'A74', 'A75' ], ['Unemployed', '1', 'Less4', 'Less7', 'More7'])
```

```python
loans['Marital_Status_Gender'] = loans['Marital_Status_Gender'].replace(['A91', 'A92', 'A93', 'A94', 'A95' ], ['MaleDivorced', 'FemaleDivOrM', 'MaleSingle', 'MaleMarriedOrWidowed', 'FemaleSingle'])
```

```python
loans['Other_Debtors_Guarantors'] = loans['Other_Debtors_Guarantors'].replace(['A101', 'A102', 'A103'], ['No', 'CoApp', 'Guarantor'])
```

```python
loans['Property'] = loans['Property'].replace(['A121', 'A122', 'A123', 'A124' ], ['RealEstate', 'LifeInsurance', 'Car', 'None'])
```

```python
loans['Other_Inst_Plans'] = loans['Other_Inst_Plans'].replace(['A141', 'A142', 'A143'], ['Bank', 'Store', 'None'])
```

```python
loans['Job'] = loans['Job'].replace(['A171', 'A172', 'A173', 'A174' ], ['NoSkills_NoRes', 'NoSkills_Res', 'LowSkills', 'HighSkills'])
```

```python
loans['Telephone'] = loans['Telephone'].replace(['A191', 'A192'], ['No', 'Yes'])
```

```python
loans['Foreign_Worker'] = loans['Foreign_Worker'].replace(['A201', 'A202'], ['Yes', 'No'])
```

```python
loans['Default_On_Payment'] = loans['Default_On_Payment'].astype(int).astype(str)
loans['Default_On_Payment']
```

```python
loans['Default_On_Payment'] = loans['Default_On_Payment'].replace(['1', '0'], ['Yes', 'No'])
```

```python
loans['Housing'] = loans['Housing'].replace(['A151', 'A152', 'A153'], ['Rent', 'Own', 'Free'])
```
